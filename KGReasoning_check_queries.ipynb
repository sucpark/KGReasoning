```python
def decode_query(query, answers):
    temp = defaultdict(lambda: defaultdict(set))
    for key in query.keys():
        if key == ('e', ('r',)):
            print(f"decoding {key} queries")
            temp_queries = query[key]
            for temp_query in temp_queries:
                temp_ent = temp_query[0]
                temp_rel = temp_query[1][0]
                decoded_query = (id2ent[temp_ent], (id2rel[temp_rel]))
                temp_answers = {id2ent[i] for i in answers[temp_query]}
                temp[key][decoded_query] = temp_answers
        elif key == (('e', ('r',)), ('e', ('r',))):
            print(f"encoding {key} queries")
            temp_queries = query[key]
            for temp_query in temp_queries:
                temp_ent1 = temp_query[0][0]
                temp_rel1 = temp_query[0][1][0]
                temp_ent2 = temp_query[1][0]
                temp_rel2 = temp_query[1][1][0]
                decoded_query = ((id2ent[temp_ent1],(id2rel[temp_rel1],)), (id2ent[temp_ent2],(id2rel[temp_rel2],)))
                temp_answers = {id2ent[i] for i in answers[temp_query]}
                temp[key][decoded_query] = temp_answers
        elif key == ((('e', ('r',)), ('e', ('r',)), ('u',)), ('r',)):
            print(f"encoding {key} queries")
            temp_queries = query[key]
            for temp_query in temp_queries:
                temp_ent1 = temp_query[0][0][0]
                temp_rel1 = temp_query[0][0][1][0]
                temp_ent2 = temp_query[0][1][0]
                temp_rel2 = temp_query[0][1][1][0]
                union = 'union'
                temp_rel3 = temp_query[1][0]
                decoded_query = (((id2ent[temp_ent1], (id2rel[temp_rel1])), (id2ent[temp_ent2], (id2rel[temp_rel2])), (union,)), (id2rel[temp_rel3],))
                temp_answers = {id2ent[i] for i in answers[temp_query]}
                temp[key][decoded_query] = temp_answers
    return temp
```
